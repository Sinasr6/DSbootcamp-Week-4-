```python
import pandas as pd # Reading datasets
import numpy as np # Operations on our dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"F:\Courses\Data Science bootcamp\archive\craigslist_vehicles.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = {"county", "id", "Unnamed: 0"}, inplace = True)
```

```python
for x in df:
    if df[x].isnull().sum() != 0:
        if df[x].dtype == "object":
                df[x] = df[x].fillna(df[x].mode().iloc[0])
        else:
            df[x] = df[x].fillna(df[x].median())
```

```python
df.isnull().sum()
```

```python
df["posting_date"] = pd.to_datetime(df["posting_date"])
df["removal_date"] = pd.to_datetime(df["removal_date"])
```

```python
df.set_index("posting_date", inplace = True)
```

```python
df.index.describe()
```

```python
df.to_csv(r"F:\Courses\Data Science bootcamp\archive\craigslist_vehicles_cleaned.csv")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Remove duplicate rows
df = df[~df.index.duplicated(keep='first')]

# Set the frequency of the index
df = df.asfreq('D')  # Replace 'D' with the appropriate frequency

# Perform seasonal decomposition
decomposition = seasonal_decompose(df['price'], model='additive')

# Plot the decomposition
fig = decomposition.plot()
plt.show()
```

```python
df.head()
```

```python
df['price'].plot()
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Price over time')
plt.show()
```

```python
df_monthly = df.resample('M').mean()  # Resample the data to monthly frequency and calculate average price
heatmap_data = df_monthly.pivot_table(index=df_monthly.index.month, columns=df_monthly.index.year, values='price')

sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Price Heatmap')
plt.show()
```
